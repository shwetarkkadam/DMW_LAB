```python
# Import libraries
import numpy as np
import pandas as pd
```

```python
# Read data from dataset into dataframe 
df = pd.read_csv('DMart.csv')
```

```python
# Display first 5 rows
df.head()
```

```python
# Get information about dataframe
df.info()
```

```python
# Data Reduction
# Remove/reduce unnecessary attributes from dataframe
subset = df[['Name', 'Brand', 'Price', 'Category', 'Quantity', 'Description']]
subset.head()
```

```python
# Data Cleaning
# Find missing values
subset.isnull().sum()
```

```python
# Define function for median
def impute_median(series):
    return series.fillna(series.median())
```

```python


subset.Price=df['Price'].transform(impute_median)
subset.Category=df['DiscountedPrice'].transform(impute_median)
```

```python
subset.isnull().sum()
```

```python
# Fill missing value of integer column with mode
subset['Brand'] = subset['Brand'].fillna(subset['Brand'].mode()[0])
subset['Category'] = subset['Category'].fillna(subset['Category'].mode()[0])
subset['Description'] = subset['Description'].fillna(subset['Description'].mode()[0])
subset['Name'] = subset['Name'].fillna(subset['Name'].mode()[0])
subset['Price'] = subset['Price'].fillna(subset['Price'].mode()[0])
subset['Quantity'] = subset['Quantity'].fillna(subset['Quantity'].mode()[0])
```

```python
subset.isnull().sum()
```

```python
# Data Reduction 
subset1 = df[['Name', 'Brand', 'Price', 'Category']]
subset1.head()
```

```python
subset2 = df[['Category', 'Quantity', 'Description']]
subset2.head()
```

```python
display(subset,subset1)
```

```python
# Data Integration
# Integrate/merge two dataframe
subset2 = pd.merge(subset,subset1,left_index=True, right_index=True)
```
